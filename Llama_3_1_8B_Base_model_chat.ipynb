```python
import transformers
import torch

model_id = "meta-llama/Meta-Llama-3.1-8B-Instruct"

pipeline = transformers.pipeline(
    "text-generation",
    model=model_id,
    model_kwargs={"torch_dtype": torch.bfloat16},
    device_map="auto",
)

messages = [
    {"role": "system", "content": "You are a pirate chatbot who always responds in pirate speak!"},
    {"role": "user", "content": "Who are you?"},
]

outputs = pipeline(
    messages,
    max_new_tokens=256,
)
print(outputs[0]["generated_text"][-1])
```

Pipeline’daki parametreler NLP konusu, ama kısaca özet geçeyim:

`max_length` (int) → Üretilen metnin maksimum uzunluğunu belirler. Bu değer, giriş metni uzunluğu dahil toplam çıktı karakter sayısını ifade eder. Dolayısıyla 50 karakterlik bir input text verip de 20 karakterlik çıktı üret diyemezsiniz

`num_return_sequences` (int) → Modelin döndüreceği bağımsız metin sayısını belirler. Bu örnekte 3 çıktı üretiyor

`do_sample` (bool) → Metin üretimi sırasında örnekleme yapılıp yapılmayacağını belirler. Eğer `True` ise, model rastgele örnekleme yapar; `False` ise, greedy çözümleme yapar.

`top_k` (int) → Her adımda olasılık dağılımından en yüksek olasılığa sahip k adet sonraki kelimeyi seçer ve bu kelimelerden birini rastgele seçimle döndürür.

`top_p` (float) → Nucleus (top-p) örnekleme, toplam olasılığı p veya daha fazla olan en küçük kelimeler setinden örnekleme yapar. Bu yöntem, örnekleme çeşitliliğini korurken mantıklı metin üretimi sağlar.

`temperature` (float) → Örnekleme sıcaklığı, üretim sürecinde çeşitliliği kontrol etmek için kullanılır. Düşük sıcaklık daha tutarlı metinler üretir, yüksek sıcaklık daha fazla çeşitlilik ve sürpriz içeren metinler üretir. Ben bunu ısındıkça fokurdamaya başlayan suya benzetirim, sıcaklık yükseldikçe sıçraması artar.

```python
messages = [
    {"role": "system", "content": "You are a helpful asistant!"},
    {"role": "user", "content": "Osmanlının ilk padişahı kimdir?"},
]

outputs = pipeline(
    messages,
    max_new_tokens=512,
    #max_length=50,
    num_return_sequences=3,
    do_sample=True,
    top_k=50,
    top_p=0.95,
    temperature=0.5 )
print(outputs[0]["generated_text"][-1]["content"])
```
